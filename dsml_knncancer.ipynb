```python
import pandas as pd
can = pd.read_csv('sample_data/data.csv')
can.head(10)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import pandas as pd
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
can['diagnosis']=le.fit_transform(can['diagnosis'])
x= can.iloc[:,:18].values
y = can['diagnosis'].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=4)
c_knn = KNeighborsClassifier(n_neighbors=4)
c_knn.fit(x_train,y_train)
y_pred = c_knn.predict(x_test)
print("Accuracy : ",metrics.accuracy_score(y_test,y_pred))
sample = [[17.99,	10.38	,122.8,	1001.0	,0.1184	,0.2776	,0.3001,	0.1471,	0.2419,	0.07871,	1.095,	0.9053,	8.589,	153.4,	0.006399,	0.04904,	0.05373	,0.01587]]
pred = c_knn.predict(sample)
pred_v = [can.diagnosis[p] for p in pred]

if pred_v==[1]:
   print("M")
else:
   print("B")
```
